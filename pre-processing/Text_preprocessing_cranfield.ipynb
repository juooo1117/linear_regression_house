```python
# 삭제 가능한 문장부호들
string.punctuation
```

```python
# tokenization(remove punctuation, numbers)
lowertext = ''.join(i for i in lowertext if not i.isdigit())  # 숫자가 아니면 string으로 저장
tokenedtext = lowertext.split()
tokenedtext2 = [x for x in tokenedtext if x not in string.punctuation]  # 특수문자들을 제거. ex) << ! : 제거O >> BUT << !!!!! : 제거X >>
tokenedtext2
```

```python
# 하지만, 일반적인 문서에는 문장부호가 단어에 붙어있음(문장부호와 문장사이에 공백 X) => 이러한 경우에는 제거가 어려움.
# re package 사용! => 문장부호, 특수문자 제거와 tokenize를 동시에 진행
import re

lowertext = ''.join(i for i in lowertext if not i.isdigit())
tokenedtext = re.findall(r"[\w']+|[.,!?;]", lowertext)
tokenedtext
```

```python
# '\W+'이용 : 알파벳이나 숫자가 아닌 모든 character (즉, punctuation)
lowertext = ''.join(i for i in lowertext if not i.isdigit())

lowertext = re.sub('\W+',' ', lowertext)
tokenedtext = lowertext.split()
tokenedtext
```

```python
import nltk
from nltk.corpus import stopwords
 
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
# remove stopwords
## 'tokenedtext'를 돌면서 stopwords가 아니면 'tokenedtext3'에 저장
stopWords = stopwords.words('english')
tokenedtext3 = [x for x in tokenedtext if x not in stopWords]
tokenedtext3
```

```python
# step5 : stemming(어간추출)
ps = PorterStemmer()
stemmedtext = [ps.stem(x) for x in tokenedtext3]

print(len(stemmedtext), len(set(stemmedtext))) # stemming 후에 unique한 단어의 개수: 69개
```

# pre-processing for whole Docs 

```python
import os
dir = '/Users/juhyeon/python-workspace/cranfieldDocs/'   # 문서가 존재하는 files path

# 해당 폴더의 파일이름들의 list를 반환 - cranfield파일 101개 
filenames = os.listdir(dir)
print(len(filenames))
```

```python
docpath = []
for filename in os.listdir(dir):
    path = (os.path.join(dir, filename))
    docpath.append(path)
    
print(len(docpath))
```

```python
type(docpath)
```

```python
def pre_processing(file_path):

    doc = ""
    for line in open(file_path, 'r', encoding='windows-1252'):            # file_path에 해당하는 파일을 읽고, 한 줄씩 string처리 
        doc += line

    # pre-processing
    cleantext = BeautifulSoup(doc, 'html5lib')                  ## step1: <TEXT> 태그를 찾아서, 안의 내용만 가져옴
    onlytext = cleantext.find('text').text 
    
    lowertext = onlytext.lower()                                ## step2: Convert to lowercase        
    str_text = ''.join(i for i in lowertext if not i.isdigit()) ## step3: remove digit
    str_text = re.sub('\W+',' ', str_text)                      ## step4: remove punctuation 
    tokenedtext = str_text.split()
    result = [x for x in tokenedtext if x not in stopWords]     ## step5: remove stopwords
    
    return result
```

```python
# pre-processing 완료된 파일 저장하는 list
whole_doc = []

for doc in docpath:
    token = pre_processing(doc)
    whole_doc.append(token)
    print(f'{doc} 파일 finish!')
```
